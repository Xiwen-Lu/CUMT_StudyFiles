# 类别转换，并建立索引
对所有二级类别，使用中文拼音代替原中文作为类别名称
具体操作思路：读取原classify_rule.json，读取每一个二级类别，然后等待用户（我）输入其英文名称，自动建立编号索引

```python
import json
import os
```

```python
def translate_ch2en(input_json_name,output_json_name,output_categories_dict,ch2en_dict_name):
    output_dict = {}
    cate_id_dict = {}
    ch2en_dict = {}
    with open(input_json_name,"rb") as f:
        data = json.load(f)
        total_cate_id = 0
        for category_level_1 in data:
            category_level_1_en = input("请输入1级类别\"{}\"的英文翻译：".format(category_level_1))
            cate_num = len(data[category_level_1])
            print("{}——{} 的二级子类数目为：{}".format(category_level_1,category_level_1_en,cate_num))
            # 二级类别的列表
            categories_en = []
            for id in range(cate_num):
                category_level_2 = data[category_level_1][id]
                category_level_2_en = input("请输入2级类别\"{}\"的英文翻译：".format(category_level_2))
                categories_en.append(category_level_2_en)
                total_cate_id += 1
                print("{} 下二级子类：{}——{}, id = {}".format(category_level_1,category_level_2,category_level_2_en,total_cate_id))
                ch2en_dict[category_level_2] = category_level_2_en
                cate_id_dict[category_level_2_en] = total_cate_id
                
            output_dict[category_level_1_en] = categories_en
            
    with open(output_json_name,"w") as f:
        json.dump(output_dict,f,indent=4)
        
    with open(output_categories_dict,"w") as f:
        json.dump(cate_id_dict,f)
        
    with open(ch2en_dict_name,"w") as f:
        json.dump(ch2en_dict,f)
```

```python
translate_ch2en("classify_rule.json","classify_rule_en.json","categories_dict.json","ch2en_dict.json")
```
